```python
def etl_data(endpoint,table_name, cookies):
    r = requests.get(url = endpoint + f'/{table_name}?format=json&Type=1&top=100&skip=0', cookies=cookies)    
    res = r.json()['results']
    df = pd.DataFrame(res)
    df = df.where(pd.notnull(df), None)
    for col in df.columns:
        if df[col].dtype == 'object':
            df[col] = convert_empty_to_none(df[col])
    df = df.applymap(str)
    if table_name == 'partner' and 'Description' not in df.columns:
        df['Description'] = None
    df.to_sql(name = f"{table_name}", con = mysql_con, schema = 'cosalon', if_exists = 'replace', index=False)
    
    i = 1
    while df.shape[0] == 100:
        r = requests.get(url = endpoint + f'/{table_name}?format=json&Type=1&top=100&skip={i * 100}', cookies=cookies)    
        res = r.json()['results']
        df = pd.DataFrame(res)
        df = df.where(pd.notnull(df), None)
        for col in df.columns:
            if df[col].dtype == 'object':
                df[col] = convert_empty_to_none(df[col])
        df = df.applymap(str)
        df.to_sql(name = f"{table_name}", con = mysql_con, schema = 'cosalon', if_exists = 'append', index=False)
        i = i + 1    
    
    return 'DONE'
```

```python
def check_ETL_Result(endpoint, table_name, cookies):
    
    r = requests.get(url = endpoint + f'/{table_name}?format=json&Type=1', cookies = cookies)
    
    actual = r.json()['__count']
    
    check_sql = f'select count(1) num_records from temp.cosalon_{table_name}'
    res = pd.read_sql(check_sql, con = mysql_con)
    res = res.num_records.iloc[0]
    return res, actual
```

```python
def etl_get_order_table(endpoint, cookies):
    
    r = requests.get(url = endpoint + '/{orders}?format=json', cookies=cookies)
    
    res = r.json()['results']
    
    df = pd.DataFrame(res)
    df = df.where(pd.notnull(df), None)
    for col in df.columns:
        if df[col].dtype == 'object':
            df[col] = convert_empty_to_none(df[col])

    df.to_sql(name = 'orders_test', con = mysql_con, schema = 'temp', if_exists = 'replace', index=False)
    return 'DONE'
```
