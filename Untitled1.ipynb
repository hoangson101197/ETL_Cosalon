```python
def check_ETL_Result(endpoint, table_name, cookies):
    
    r = requests.get(url = endpoint + f'/{table_name}?format=json&Type=1', cookies = cookies)
    
    actual = r.json()['__count']
    
    check_sql = f'select count(1) num_records from cosalon.{table_name}'
    res = pd.read_sql(check_sql, con = mysql_con)
    res = res.num_records.iloc[0]
    return res, actual
```
